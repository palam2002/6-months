```python
fifa_df = pd.read_csv("/Users/palammysurareddy/Downloads/14th- Seaborn, Eda practicle/Seaborn/FIFA.csv").head(50)
fifa_df
```

```python
f, ax = plt.subplots(figsize=(8,6))
x = fifa_df['Age']
ax = sns.distplot(x, bins=10)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,6))
x=fifa_df['Age']
ax=sns.distplot(x,bins=10,vertical=True)
```

```python
f, ax = plt.subplots(figsize=(8,6))
x = fifa_df['Age']
x = pd.Series(x, name="Age variable")
ax = sns.kdeplot(x)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,6))
x = fifa_df['Age']
x = pd.Series(x, name="Age variable")
ax = sns.kdeplot(x,color='g',shade=True)
plt.show()
```

```python
sns.distplot(fifa_df['Age'],bins=10,hist=False)
plt.grid()
```

```python
f,ax=plt.subplots(figsize=(6,7))
sns.countplot(x='Preferred Foot',data=fifa_df,color='y')
plt.grid()
```

```python
#Visualize distribution of values with Seaborn countplot() function.
f,ax=plt.subplots(figsize=(5,3))
sns.countplot(x='Preferred Foot',hue='Real Face',data=fifa_df,color='c')
plt.grid()
```
