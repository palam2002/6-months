```python
import transformers
import torch

model_id = "meta-llama/Llama-3.1-8B"

# Authenticate to Hugging Face if you haven't already
!huggingface-cli login

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto"
)

pipeline("who is captain of india?")
```

```python
pipeline("Who is the captain of Indian cricket team")
```
