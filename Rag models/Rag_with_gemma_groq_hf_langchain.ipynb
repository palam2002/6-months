```python
# Document Loader
from langchain.document_loaders import TextLoader
loader = TextLoader('/content/drive/MyDrive/yolo fime  names.txt')
documents = loader.load()
```

```python
import textwrap

def wrap_text_preserve_newlines(text, width=110):
    lines = text.split('\n')
    wrapped_lines = [
        textwrap.fill(line, width=width) if line.strip() else ""
        for line in lines
    ]
    return '\n'.join(wrapped_lines)
```

```python
print(wrap_text_preserve_newlines(str(documents[0])))
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
len(docs)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

```python
from langchain.vectorstores import FAISS

db = FAISS.from_documents(docs, embeddings)
```

```python
query = "what about rag file"
docs = db.similarity_search(query)
```

```python
print(wrap_text_preserve_newlines(str(docs[0].page_content)))
```

```python
import langchain_groq
from langchain_groq import ChatGroq

GROQ_LLM = ChatGroq(
            api_key=os.getenv("GROQ_API_KEY"),
            model="gemma2-9b-it"
        )
```

```python
from langchain.chains.question_answering import load_qa_chain

chain = load_qa_chain(GROQ_LLM, chain_type="stuff")
```

```python
query = "HOW IS INDIAN ECONOMY"
docs = db.similarity_search(query)
chain.run(input_documents=docs, question=query)
```
